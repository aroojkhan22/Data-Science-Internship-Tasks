# Task 1: Exploring Iris Dataset

This task aims to explore and visualize the Iris dataset using Python libraries like pandas, matplotlib, and seaborn.


```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('iris')
df.head()
```

## Dataset Overview

The dataset contains information about iris flowers.
It includes measurements of sepal and petal along with species classification.

## Data Understanding

The dataset has 150 rows and 5 columns.
There are no missing values.
Four columns are numerical and one column (species) is categorical.

## Visualization

Now we will create graphs to understand relationships and distribution in the dataset.

```python
import matplotlib.pyplot as plt

plt.scatter(df['sepal_length'], df['sepal_width'])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Width")
plt.show()
```

This scatter plot shows the relationship between sepal length and sepal width.

```python
df['petal_length'].hist()
plt.title("Petal Length Distribution")
plt.show()
```

The histogram shows how petal length values are distributed across the dataset.

```python
import seaborn as sns

sns.boxplot(x=df['sepal_length'])
plt.title("Boxplot of Sepal Length")
plt.show()
```

The boxplot helps identify outliers and shows the spread of sepal length values.
